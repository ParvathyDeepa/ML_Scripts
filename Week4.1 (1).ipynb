```python
#Question 1

from sklearn.datasets import load_diabetes
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load the diabetes dataset
diabetes = load_diabetes()
data = diabetes.data  # Features
target = diabetes.target  # Target variable (optional, not used for PCA)

# Apply PCA with 3 components
pca = PCA(n_components=3)
pca_data = pca.fit_transform(data)

# Extract the first three principal components
pc1 = pca_data[:, 0]
pc2 = pca_data[:, 1]
pc3 = pca_data[:, 2]

# Plot the first three components
plt.scatter(pc1, pc2, label='PC1 vs PC2')
plt.scatter(pc1, pc3, label='PC1 vs PC3')
plt.scatter(pc2, pc3, label='PC2 vs PC3')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2 (or 3)')
plt.title('PCA of Diabetes Dataset (First Three Components)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Question 2

from sklearn.datasets import load_diabetes
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load the diabetes dataset
diabetes = load_diabetes()
data = diabetes.data  # Features

# Apply PCA with 3 components
pca = PCA(n_components=3)
pca_data = pca.fit_transform(data)

# Explained variance ratio for each component
explained_variance = pca.explained_variance_ratio_

# Total variance
total_variance = explained_variance.sum()

# Explained variance ratio for the first 3 components (as percentages)
explained_variance_ratio_3 = explained_variance[:3] * 100

print("Explained Variance Ratio:")
for i, ratio in enumerate(explained_variance_ratio_3):
    print(f"  - PC{i+1}: {ratio:.2f}%")

# Rest of the plotting code (same as before)
plt.scatter(pc1, pc2, label='PC1 vs PC2')
plt.scatter(pc1, pc3, label='PC1 vs PC3')
plt.scatter(pc2, pc3, label='PC2 vs PC3')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2 (or 3)')
plt.title('PCA of Diabetes Dataset (First Three Components)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Question 4

from sklearn.datasets import load_diabetes
from sklearn.manifold import TSNE  # Import TSNE
import matplotlib.pyplot as plt

# Load the diabetes dataset
diabetes = load_diabetes()
data = diabetes.data  # Features

# Apply t-SNE with 3 components (method='exact')
tsne = TSNE(n_components=3, method='exact')  # Specify method='exact'
tsne_data = tsne.fit_transform(data)

# Extract the first three components
tsne_pc1 = tsne_data[:, 0]
tsne_pc2 = tsne_data[:, 1]
tsne_pc3 = tsne_data[:, 2]

# Plot the first three components
plt.scatter(tsne_pc1, tsne_pc2, label='t-SNE PC1 vs PC2')
plt.scatter(tsne_pc1, tsne_pc3, label='t-SNE PC1 vs PC3')
plt.scatter(tsne_pc2, tsne_pc3, label='t-SNE PC2 vs PC3')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2 (or 3)')
plt.title('t-SNE of Diabetes Dataset (First Three Components)')
plt.legend()
plt.grid(True)
plt.show()
```
